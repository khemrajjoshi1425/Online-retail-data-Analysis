# Importing Necessary Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from datetime import datetime
```

# Loading Dataset csv file

```python
online_retail_df = pd.read_csv('/content/retail.csv', parse_dates=['InvoiceDate'], encoding='latin1')
online_retail_df.head()
```

```python

online_retail_df.shape
```

```python
online_retail_df.info()
```

```python
# converting the invoiceDate to python readable datetime object
online_retail_df["InvoiceDate"] = pd.to_datetime(
    online_retail_df["InvoiceDate"], format="%m/%d/%y %H:%M"
)
online_retail_df.head()
```

```python
online_retail_df.info()
```

# Extracting all Fourth month data from dataset

```python
fourth_month_df = online_retail_df[online_retail_df["InvoiceDate"].dt.month == 4]
fourth_month_df
```

# Extracting all Tenth month data from dataset

```python
tenth_month_df = online_retail_df[online_retail_df["InvoiceDate"].dt.month == 10]
tenth_month_df
```

# Analysis the Box plot for the Customer Purchase price for Month 4

```python
fourth_month_df["customer_purchase_price"] = fourth_month_df["Quantity"] * fourth_month_df["UnitPrice"]
fourth_month_df
```

```python
customer_totals_purchase = (
    fourth_month_df[fourth_month_df["Quantity"] >= 0]
    .groupby("CustomerID")["customer_purchase_price"]
    .sum()
    .reset_index()
)
customer_totals_purchase
```

```python
customer_totals_purchase["customer_purchase_price"].describe()
```

```python
fig, axs = plt.subplots(1, 1, figsize=(20, 3))
plt1 = sns.boxplot(
    data=customer_totals_purchase, x="customer_purchase_price", ax=axs
)
```

# Analysing the Box plot for the Hours of transaction by a customer for Month 4

```python
hours_of_transaction = (
    fourth_month_df.groupby("CustomerID")["InvoiceDate"]
    .apply(lambda x: x.dt.hour.sum())
).reset_index()
hours_of_transaction.columns = ["CustomerID", "Hours_of_Transaction"]
hours_of_transaction
```

```python
fig, axs = plt.subplots(1, 1, figsize=(20, 3))
plt2 = sns.boxplot(data=hours_of_transaction,x="Hours_of_Transaction", ax=axs)
```

# Sales Analysis for Weekdays and Weekends for Month 4

```python
fourth_month_df["DayOfWeek"] = fourth_month_df["InvoiceDate"].dt.day_name()

fourth_month_df["DayType"] = fourth_month_df["InvoiceDate"].dt.dayofweek.apply(
    lambda x: "Weekend" if x >= 5 else "Weekday"
)

sales_by_daytype = fourth_month_df.groupby("DayType")["Quantity"].sum()

print(sales_by_daytype)

fig, axs = plt.subplots(1, 1, figsize=(10, 5))
plt5 = sns.barplot(data=sales_by_daytype, ax=axs)
```

# Ranking Customers in Month 4 based on the Quantity Purchased

```python
customer_rank_month_4 = (
    fourth_month_df[fourth_month_df["Quantity"] >= 0]
    .groupby("CustomerID")["Quantity"]
    .sum()
    .reset_index()
)
print(customer_rank_month_4)

customer_rank_month_4["Rank"] = customer_rank_month_4["Quantity"].rank(
    ascending=False, method="min"
)

customer_rank_month_4 = customer_rank_month_4.sort_values(by="Rank")

customer_rank_month_4
```

```python
customer_rank_count_month_4 = (
    customer_rank_month_4["Rank"].value_counts().reset_index()
)
customer_rank_count_month_4.columns = ["Rank", "Customer_Count"]

# Sort ranks by ascending order
customer_rank_count_month_4 = customer_rank_count_month_4.sort_values(by="Rank")
customer_rank_count_month_4
```

```python
fig, axs = plt.subplots(1, 1, figsize=(20, 8))
plt5 = sns.lineplot(x="Rank", y="Customer_Count", data=customer_rank_count_month_4, ax=axs)
```
